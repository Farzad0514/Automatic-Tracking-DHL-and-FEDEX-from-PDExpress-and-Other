```python
from selenium import webdriver
import time
driver = webdriver.Firefox(executable_path="C:/Users/Farzad/webDrivers/geckodriver.exe")
#url= input('URL : ')
url='http://www.pdexpress.net'
driver.get(url)
```

```python
import pandas as pd
import time
from random import randint
from time import sleep
data= pd.read_csv('trackingcode/tracking.csv')

dhlTrackingCodes = data['tracking']

#dhlTrackingCodes = data['fedex']

len(dhlTrackingCodes)

l=[]

for k in range(0, int(len(dhlTrackingCodes))):
#for k in range(0, 10):
    
    d={}
    
    try:
        d['Tracking']= str(dhlTrackingCodes[k])
    
    except:
        d['Tracking']= 'invalid'

    try:
        url='http://www.pdexpress.net'
        driver.get(url)

        sleep(randint(2,7))

        dropDown= driver.find_element_by_xpath('//*[@id="select"]').click()
        option= driver.find_elements_by_tag_name('option')

        if len(str(dhlTrackingCodes[k])) ==10:

            option[9].click() #select DHL 
        #option[0].click() #select FedEx
            d['Tracking validity']= 'Valid'

        if len(str(dhlTrackingCodes[k])) ==12:
            option[0].click() #select FedEx
            d['Tracking validity']= 'Valid'
            
        inputField= driver.find_element_by_xpath('//*[@id="input"]')
        inputField.send_keys(int(dhlTrackingCodes[k])) #enter tracking number
        btn= driver.find_element_by_css_selector('body > table > tbody > tr:nth-child(2) > td > table > tbody > tr:nth-child(2) > td > form > button').click() #click track button


        try:
            table= driver.find_element_by_class_name('text1')
            tbody= table.find_element_by_tag_name('tbody')
            tr= tbody.find_elements_by_tag_name('tr')

            d['Fetching status'] = 'Success'


            for i in range (0, int(len(tr))):

                td= tr[i].find_elements_by_tag_name('td')

                if len(str(dhlTrackingCodes[k])) ==10:

                    try:
                        if 'To' in td[0].text:
                            d['Country']=td[1].text
                    except: d['Country']=None

                    try:
                        if 'Shipper name' in td[0].text:
                            d['Shipper'] =td[1].text
                    except: d['Shipper'] =None

                    try:
                        if 'Shipper ref' in td[0].text:     
                            d['Apx No'] =td[1].text
                    except: d['Apx No'] = None

                    try:
                        if 'desc' in td[0].text:     
                            d['Description'] =td[1].text
                    except: d['Description'] =None

                    try:
                        if 'date' in td[0].text:     
                            d['Date'] =td[1].text
                    except: d['Date'] = None

                    try:
                        if 'Consignee' in td[0].text:     
                            d['CONSIGNEE NAME'] =td[1].text
                    except:
                         d['CONSIGNEE NAME'] =None

                    try:
                        if 'Weight' in td[0].text:     
                            d['Weight Charged'] =td[1].text

                    except:
                        d['Weight Charged'] =None

                    try:
                        if 'status' in td[0].text:     
                            d['Status'] =td[1].text

                    except:

                         d['Status'] =None

                if len(str(dhlTrackingCodes[k])) ==12:


                    try:
                        if 'Destination' in td[0].text:
                            d['Destination']=td[1].text

                    except:
                        d['Destination']=None

                    try:
                        if 'Purchase order' in td[0].text:
                            d['Purchase order'] =td[1].text     
                    except:
                        d['Purchase order'] =None

                    try:
                        if 'Reference' in td[0].text:     
                            d['Apx No'] =td[1].text
                    except:
                        d['Apx No'] = None

                    try:
                        if 'desc' in td[0].text:     
                            d['Description'] =td[1].text

                    except:
                        d['Description'] = None


                    try:
                        if 'date' in td[0].text:     
                            d['Date'] =td[1].text

                    except:
                        d['Date'] = None

                    try:

                        if 'Consignee' in td[0].text:     
                            d['CONSIGNEE NAME'] =td[1].text

                    except:
                         d['CONSIGNEE NAME'] = None

                    try:
                        if 'Package weight' in td[0].text:     
                            d['Weight Charged'] =td[1].text

                    except:
                        d['Weight Charged'] = None

            
                        

        except: 
            print('Failed for: ' + str(dhlTrackingCodes[k]))
            d['Fetching status'] = 'failed'

    except: 
        print('Failed for at start: ' + str(dhlTrackingCodes[k]))
        d['Tracking validity']= 'Invalid'

    #sleep(randint(2,7))
    l.append(d)
        #driver.close()
```

```python
df= pd.DataFrame(l)
```
